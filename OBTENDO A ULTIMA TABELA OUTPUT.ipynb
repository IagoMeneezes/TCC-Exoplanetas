```python
import pandas as pd


df = pd.read_csv('C:/Users/Iagou/OneDrive/Área de Trabalho/UFS/TCC/Python/ULTIMA TABELA INPUT 29-10.csv',delimiter=',',low_memory = False)


NomeStar = df['Nome da estrela']
NomeExo = df['Nome do planeta']
ExoMT = df['PMterrestres']
ExoRT = df['PRterrestres']
PeriodoDias = df['PPeriodoDIAS']
MassaStar = df['Massa da estrela']
TempStar = df['TemperaturaStar']
RaioStar = df['RaioStar']
AmpVeloRad = df['Radial Velocity Amplitude']
Transit = df['Profundidade de transito']
Excentric = df['Excentricidade']

#df

#------------------------------------------------------CONSTANTES---------------------------------------------------------


# Unidade Astronômica 
UA = 149597870700               #Em metros

# VALOR DE PI
pi = 3.141592654

# CONSTANTE GRAVITACIONAL UNIVERSAL (COM UNIDADES m³ kg-¹ s-²)
G = 6.674184*(10**-11)  

seni = 1

# Massa da Terra e Júpiter
RaioTerra = 6371000                #EM METROS
RaioJupiter = 69911000     # EM METROS
MassaTerra = 5.972*(10**24)     #Em KG
MassaJupiter = 1.898*(10**27)   #Em KG

# Dados do Sol

Tsol = 5778               # Temperatura do sol em Kelvin
Rsol = (6.960)*(10**8)      # Raio do Sol em Metros
Lsol = (3.846)*(10**33)   # Luminosidade do Sol  (ERG/S)
Msol = (1.9891*(10**30))  # Massa do Sol em KG
                 

# Como estamos estudando exoplanetas parecidos com a Terra, utilizaremos o Albedo da Terra  (At)
At = 0.306

# Constante de Stefan - Boltzman
ConstSB = (5.67)*(10**-8)

# RAIO DA ESTRELA (SOLAR) TRANSFORMADO EM METROS 
RStarTrans = (RaioStar)*(Rsol) 
#O VALOR DA MASSA PRECISA ESTAR TRANSFORMADO PARA MASSA SOLARES
M = MassaStar*Msol


# --------------------------------------------------------------------------------------------------------------------------


#                                       Formulas pro TCC - calculo de Parametros




#                                                   SEMI EIXO MAIOR

# Para determinar o SEMI EIXO MAIOR do exoplaneta, precisaremos utilizar a 3° lei de Kepler


#O PERIODO PRECISA SER TRANSFORMADO PRA SEGUNDOS
p = (PeriodoDias*24*60*60)
p2 = p**2

#O VALOR DA MASSA PRECISA ESTAR TRANSFORMADO PARA MASSA SOLARES
#M = MassaStar*Msol


# O semi eixo maior é chamado de 'a'
ai = (p2*G*M) / (4*(pi**2))


a = (ai)**(1/3)   #EM METROS

#print ('O valor do semi eixo maior (a) é {0} metros'.format(a))

# EM UNIDADES ASTRONOMICAS 
aUA = (a)/(UA)

#print ('O semi eixo maior em UA vale {0} UA'.format(aUA))

#vetor = estrela, nomedoplaneta, massaterr, raioterr, periodoDIAS, massastar, TempStar, RaioStar, aUA

#dataf = pd.DataFrame({'Nome da estrela': vetor[0], 'Nome do planeta': vetor[1], 'PMterrestres': vetor[2],'PRterrestres': vetor[3], 'PPeriodoDIAS': vetor [4], 'Massa da estrela': vetor [5], 'TemperaturaStar': vetor [6], 'RaioStar': vetor [7], 'SemiEixoMaior': vetor [8]})
#dataf


# --------------------------------------------------------------------------------------------------------------------------

#                                              GRAU DE INSOLAÇÃO


# Para determinar o GRAU DE INSOLAÇÃO, que é a quantidade de energia que atravessa a estrela por unidade de área por unidade
# de tempo, utilizaremos:


#Formula do grau de insolação do exoplaneta

#O RAIO DA ESTRELA PRECISA ESTAR EM METROS ... E TA EM RAIOS SOLARES, PRECISA TRANSFORMAR.

Sp = ((RStarTrans/Rsol)**2)*((TempStar/Tsol)**4)*(1/(aUA)**2)

#print ('O valor do grau de insolação do exoplaneta em função do grau de insolação terrestre é {0} St'.format(Sp))



#vetor = estrela, nomedoplaneta, massaterr, raioterr, periodoDIAS, massastar, TempStar, RaioStar, aUA, Sp

#dataf = pd.DataFrame({'Nome da estrela': vetor[0], 'Nome do planeta': vetor[1], 'PMterrestres': vetor[2],'PRterrestres': vetor[3], 'PPeriodoDIAS': vetor [4], 'Massa da estrela': vetor [5], 'TemperaturaStar': vetor [6], 'RaioStar': vetor [7], 'SemiEixoMaior': vetor [8], 'GraudeInsolação': vetor [9]})
#dataf



# --------------------------------------------------------------------------------------------------------------------------



#                                        TEMPERATURA DE EQUILIBRIO


#  Partindo da premissa que a energia absorvida pelo planeta é igual a sua energia emitida, podemos calcular a temperatura
#  de equilibrio (Teq). Temperatura a qual o corpo teria, caso, a sua única fonte de energia fosse a radiação estelar.

#  Como estamos estudando exoplanetas parecidos com a Terra, utilizaremos o Albedo da Terra  (At)


Teq = (TempStar)*((1-At)**(1/4))*((RStarTrans/(2*a))**(1/2))

#print ('O valor da temperatura de equilibrio do exoplaneta é de aproximandamente {0:.5} Kelvin'.format(Teq))


#vetor = estrela, nomedoplaneta, massaterr, raioterr, periodoDIAS, massastar, TempStar, RaioStar, aUA, Sp, Teq

#dataf = pd.DataFrame({'Nome da estrela': vetor[0], 'Nome do exoplaneta': vetor[1], 'Exoplaneta em Massas terrestres': vetor[2],'Exoplaneta em Raios terrestres': vetor[3], 'Periodo em DIAS': vetor [4], 'Massa da estrela': vetor [5], 'Temperatura da estrela': vetor [6], 'Raio da estrela': vetor [7], 'SemiEixo Maior': vetor [8], 'Grau de Insolação': vetor [9], 'Temp de Equilibrio': vetor[10]})
#dataf



# --------------------------------------------------------------------------------------------------------------------------


#                                                       DENSIDADE


# A Densidade de um exoplaneta pode ser obtida com os dados da massa e raio do planeta, pela relação:

# d = massaterr/V  , onde massaterr é a massa do exoplaneta e V é o volume. 

# O volume é obtido através da equação:  V = (4/3)*(pi)*(raioterr**3)

# Lembre que: O raio deve estar em centimetros e a massa em gramas.


Rexo = (ExoRT)*(RaioTerra)*100  #transformação para raio terrestres em Metros e depois para cm

Mexo = (ExoMT)*(MassaTerra)*1000  # transformação para massa terrestre em Kg e depois para gramas

V = (4/3)*(pi)*(Rexo**3)

d = Mexo/V

#print ('O valor da densidade do exoplaneta é de aproximadamente: {0} g/cm**3'.format(d))


#vetor = estrela, nomedoplaneta, massaterr, raioterr, periodoDIAS, massastar, TempStar, RaioStar, aUA, Sp, Teq, d

#dataf = pd.DataFrame({'Nome da estrela': vetor[0], 'Nome do exoplaneta': vetor[1], 'Exoplaneta em Massas terrestres': vetor[2],'Exoplaneta em Raios terrestres': vetor[3], 'Periodo em DIAS': vetor [4], 'Massa da estrela': vetor [5], 'Temperatura da estrela': vetor [6], 'Raio da estrela': vetor [7], 'SemiEixo Maior': vetor [8], 'Grau de Insolação': vetor [9], 'Temp de Equilibrio': vetor[10], 'Densidade (g/cm³)': vetor[11]})
#dataf




# --------------------------------------------------------------------------------------------------------------------------
#                                              ZONA HABITÁVEL

# Para o estudo da Zona Habitabilidade, é preciso saber se o planeta se encontra a uma distância, tal que, haja a possibilidade
# de existir água no estado líquido. Pois, a vida como conhecemos no nosso planeta Terra, necessita de forma fundamental que 
# isso seja satisfeito

# Para este cálculo, utilizaremos as seguintes fórmulas: 

#   Hz interna = 0.75*((Lstar/Lsol)**(1/2))
#   Hz interna = 1.0*((Lstar/Lsol)**(1/2))
#   Hz interna = 1.77*((Lstar/Lsol)**(1/2))

# Onde L é a luminosidade da estrela, obtida através da lei de Stefan-Boltzman

Lstar = (4)*(pi)*(RStarTrans**2)*(ConstSB)*(TempStar**4)  # Unidade é Watt

#print ('O valor da luminosidade da estrela é igual a: {0} Watt'.format(Lstar))

# Para transformar para erg/s precisaremos multiplicar por 10**7, obteremos: 

LstarErgs = Lstar*(10**7)

#print ('O valor da luminosidade da estrela em erg/s é igual a: {0}'.format(LstarErgs))

# Substituindo agora nas fórmulas da Hz.

Hzinterna = 0.75*((LstarErgs/Lsol)**(1/2))
Hzcentral = 1.0*((LstarErgs/Lsol)**(1/2))
Hzexterna = 1.77*((LstarErgs/Lsol)**(1/2))

#print ('A zona habitável interna começa a uma distância de aproximadamente {0} UA'.format(Hzinterna))
#print ('A zona habitável central está localizada a uma distância de aproximadamente {0}'.format(Hzcentral))
#print ('A zona habitável externa se estende até uma distância de aproximadamente {0}'.format(Hzexterna))

# PARA SABERMOS SE O EXOPLANETA SE ENCONTRA NA ZONA HABITÁVEL, ELE PRECISA ESTAR DENTRO DESSA ZONA, OU SEJA: 
# SEU SEMI EIXO MAIOR PRECISA ESTAR DENTRO DESSES LIMITES.

# --------------------------------------------------------------------------------------------------------------------------

#                  CALCULO DA MASSA DO EXOPLANETA ATRAVÉS DA AMPLITUDE DA VELOCIDADE RADIAL

#MasExoAmpVR = (AmpVeloRad)*(MassaStar**(2/3))*((1-(Excentric**2))**(1/2))*((p/(2*pi*G))**(1/3)

# TENTEI FAZER O CALCULO COM UMA ÚNICA FORMULA, MAS TA DANDO UM ERRO...
# VOU TENTAR REALIZA-LO POR PARTES 

PARTE1 = (p/(2*pi*G))**(1/3)
PARTE2 = (1-(Excentric**2))**(1/2)
PARTE3 = (M**(2/3))

MasExoAmpVR = (AmpVeloRad)*(PARTE1)*(PARTE2)*(PARTE3)

# Para transformar a massa do exoplaneta em massas Terrestres... precisamos fazer a RAZÃO

MassEAVR = (MasExoAmpVR/MassaTerra)
#print (MassEAVR)
                                                                         
                                                                         


# PARA COMPARAR AS RESPOSTAS, RESOLVI PLOTAR UMA NOVA TABELA

#VETORTESTE = NomeStar, NomeExo, ExoMT, MassEAVR
#datateste = pd.DataFrame ({'NomeStar': VETORTESTE[0],'NomeExo': VETORTESTE[1],'MassadaNASA': VETORTESTE[2],'MassaCalculada': VETORTESTE[3]})
#datateste



# --------------------------------------------------------------------------------------------------------------------------

#                         CALCULO DO RAIO DO EXOPLANETA ATRAVÉS DA PROFUNDIDADE DE TRÂNSITO


RS2 = RaioStar**2
RP = ((Transit*RS2)/1.049)**(1/2)
RPTJ = RP*RaioJupiter
RPTT = RPTJ/RaioTerra
#print (RPTT)

# PARA COMPARAR AS RESPOSTAS, RESOLVI PLOTAR UMA NOVA TABELA

#VETORTESTE1 = NomeStar, NomeExo, ExoRT, RPTT
#datateste = pd.DataFrame ({'NomeStar': VETORTESTE1[0],'NomeExo': VETORTESTE1[1],'RAIOdaNASA': VETORTESTE1[2],'RaioCalculado': VETORTESTE1[3]})
#datateste






# --------------------------------------------------------------------------------------------------------------------------


#vetorinput = NomeStar, NomeExo, PeriodoDias, MassaStar, TempStar, RaioStar, AmpVeloRad, Transit, Excentric
#dataf1 = pd.DataFrame({'Nome da estrela': vetorinput[0], 'Nome do planeta': vetorinput[1], 'PPeriodoDIAS': vetorinput [2], 'Massa da estrela': vetorinput [3], 'TemperaturaStar': vetorinput [4], 'RaioStar': vetorinput [5], 'Amplitude da Velocidade Radial': vetorinput[6], 'Profundidade de Trânsito': vetorinput[7], 'Excentricidade': vetorinput[8]})
#dataf1
#dataf1.to_csv('TabelaINPUT.csv', index = False) 

# --------------------------------------------------------------------------------------------------------------------------


#vetoroutput = NomeStar, NomeExo, MassEAVR, RPTT, Sp, Teq, d, aUA, Hzinterna, Hzcentral, Hzexterna, GoldenLuckZone
#dataf2 = pd.DataFrame({'Nome da estrela': vetoroutput[0], 'Nome do planeta': vetoroutput[1], 'Massa Exoplaneta Calculada': vetoroutput[2], 'Raio Exoplaneta Calculado': vetoroutput[3], 'Grau de Insolação': vetoroutput [4], 'Temp de Equilibrio': vetoroutput [5], 'Densidade (g/cm³)': vetoroutput [6], 'SemiEixo Maior': vetoroutput [7], 'Hzinterna': vetoroutput[8],'Hzcentral': vetoroutput[9],'Hzexterna': vetoroutput[10],'Golden LuckZone': vetoroutput[11]})

GoldenLuckZone = ['False', 'False', 'False', 'False', 'False', 'False', 'False', 'False', 'False', 'False', 'False', 'False','False', 'False', 'False', 'False', 'False', 'False','False', 'False', 'False', 'False', 'False', 'False','False', 'False', 'False', 'False', 'False', 'False','False', 'False', 'False', 'False', 'False', 'False','False', 'False', 'False', 'False', 'False', 'False','False', 'False', 'False', 'False', 'False', 'True', 'True','True', 'False', 'False']
#GoldenLuckZone = ['False', 'False', 'False','False','False','False','False','False','False','False','False','False','False','True', 'True', 'True', 'False','False','False','False','False','False','False','False','False','False','False','False','False','False','False','False','False','False','False','False','False','False','False','False','False','False']
#dataf2['Zona Habitável'] = GoldenLuckZone
#dataf2
#dataf2.to_csv('TabelaOUTPUT.csv', index = False) 

# --------------------------------------------------------------------------------------------------------------------------


vetor = NomeStar, NomeExo, ExoMT, MassEAVR, ExoRT, RPTT, PeriodoDias, MassaStar, TempStar, RaioStar, AmpVeloRad, Transit, Excentric, Sp, Teq, d, aUA, Hzinterna, Hzcentral, Hzexterna, GoldenLuckZone
dataf = pd.DataFrame({'Nome da estrela': vetor[0], 'Nome do planeta': vetor[1], 'Massa Exoplanet NASA': vetor[2],'Massa Exoplanet Calculada': vetor[3],'Raio Exoplanet NASA': vetor[4], 'Raio Exoplanet Calculado': vetor[5], 'PPeriodoDIAS': vetor [6], 'Massa da estrela': vetor [7], 'TemperaturaStar': vetor [8], 'RaioStar': vetor [9], 'Amplitude da Velocidade Radial': vetor [10], 'Profundidade de Trânsito': vetor [11], 'Excentricidade': vetor [12], 'Grau de Insolação': vetor [13], 'Temp de Equilibrio': vetor [14], 'Densidade (g/cm³)': vetor [15], 'SemiEixo Maior': vetor [16], 'Hzinterna': vetor[17],'Hzcentral': vetor[18],'Hzexterna': vetor[19], 'ZonaHabitável': vetor[20]})
dataf

#dataf.to_csv('ULTIMATabelaUnica.csv', index = False) 

#dataf.dropna(axis=0, how='any').to_csv('TABELACOMPLETA.csv', index = False) 
# .to_csv ... Salvou o arquivo em CSV com o nome escolhido em (' NOME ')
# Index = False ... retirou a posição que estava os dados na tabela original.
```
